### Parsed NVD to Dataframe

```python
import glob
import pandas as pd

path = '../data/nvd-hashs/'
allcves = glob.glob(path+'*.csv')
l = []

for filename in allcves:
    df = pd.read_csv(filename)
    l.append(df)
df_NVD_hash = pd.concat(l, ignore_index=True)
#New cell
df_NVD_hash
```

### Reading in SHGD revision.csv and cleaning

```python
import datetime as dt
import re
import matplotlib.pyplot as plt

rev_df = pd.read_csv('C:/Users/Faugh/Desktop/DataScienceData/popular-3k-python/revision.csv', header=None)
rev_df.columns = ['id', 'date', 'date_offset', 'committer_date', 'committer_date_offset', 'type', 'directory', 'message', 'author', 'committer']
rev_df = rev_df[['id', 'date', 'message']]

rev_df['id'] = rev_df['id'].apply(lambda x: x[2:])
rev_df['date'] = rev_df['date'].apply(lambda x: dt.datetime.strptime(x,'%Y-%m-%dT%H:%M:%S.000Z'))
rev_df['message'] = rev_df['message'].apply(lambda x: bytes.fromhex(x[2:]).decode(encoding='ISO-8859-1'))
rev_df
```

### Comparing NVD and SHGD
* Giving rev_df a common column name "hash" with NVD_df
* Merging rev_df and NVD_df on "hash" to see what hashes are in common between said datasets

```python
rev_df.columns = ['hash', 'date', 'message']
merged_NVD_Revision = pd.merge(df_NVD_hash, rev_df, on=['hash'], how='inner')
merged_NVD_Revision
```
