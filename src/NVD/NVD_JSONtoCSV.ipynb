```python
import json
import pandas as pd
from pandas.io.json import json_normalize 

with open("nvdcve-1.1-recent.json") as f:
    d = json.load(f)
```

CVE_Items

```python
recent = json_normalize(d["CVE_Items"])
recent
```

```python
def getID(desLength):
    for a in range(desLength):
        id = recent.iloc[a]["cve.CVE_data_meta.ID"]  
        data.loc[a,'cve.CVE_data_meta.ID'] = id
```

```python
def description(desLength):
    for a in range(desLength): 
        desLength = len(recent.iloc[a]["cve.description.description_data"])
        for x in range(desLength):
            lang = recent.iloc[a]["cve.description.description_data"][x]["lang"]
            value = recent.iloc[a]["cve.description.description_data"][x]["value"]
            data.loc[a,'desLang'] = lang
            data.loc[a,'desValue'] = value
```

```python
def problem(desLength):
    for a in range(desLength):
        proLength = len(recent.iloc[a]["cve.problemtype.problemtype_data"])
        for x in range(proLength):
            desLength = len(recent.iloc[a]["cve.problemtype.problemtype_data"][x]["description"])
            for y in range(desLength):
                lang = recent.iloc[a]["cve.problemtype.problemtype_data"][x]["description"][y]["lang"]
                value = recent.iloc[a]["cve.problemtype.problemtype_data"][x]["description"][y]["value"]
                data.loc[a,'proLang'] = lang
                data.loc[a,'proValue'] = value
```

```python
def destination(desLength):
    for a in range(desLength):
        refLength = len(recent.iloc[a]["cve.references.reference_data"])
        for x in range(refLength):
            url = recent.iloc[a]["cve.references.reference_data"][x]["url"]
            name = recent.iloc[a]["cve.references.reference_data"][x]["name"]
            refSource = recent.iloc[a]["cve.references.reference_data"][x]["refsource"]
            tags = recent.iloc[a]["cve.references.reference_data"][x]["tags"]
            data.loc[a,'refUrl'] = url
            data.loc[a,'refName'] = name
            data.loc[a,'refSource'] = refSource
            data.loc[a,'refTags'] = tags
```

```python
def config(desLength):
    for a in range(desLength):
        conLength = len(recent.iloc[a]["configurations.nodes"])
        for x in range(conLength):
            operator = recent.iloc[a]["configurations.nodes"][x]["operator"]
            if (len(recent.iloc[a]["configurations.nodes"][x]["cpe_match"]) == 0):
                print("hi")
            cpeLength = len(recent.iloc[a]["configurations.nodes"][x]["cpe_match"])
            for y in range(cpeLength):
                vulnerable = recent.iloc[a]["configurations.nodes"][x]["cpe_match"][y]["vulnerable"]
                cpe23Uri = recent.iloc[a]["configurations.nodes"][x]["cpe_match"][y]["cpe23Uri"]
                #versionEndExcluding = recent.iloc[a]["configurations.nodes"][x]["cpe_match"][y]["versionEndExcluding"]
                data.loc[a,'conVul'] = vulnerable
                data.loc[a,'conCpe'] = cpe23Uri
                #data.loc[a,'ConVer'] = versionEndExcluding
```

```python
data = pd.DataFrame(columns=['cve.CVE_data_meta.ID','desLang','desValue','proLang','proValue','refUrl','refName','refSource','refTags','conOper','conVul',"conCpe","conVer"])
desLength = recent.shape[0]

getID(desLength)
description(desLength)
problem(desLength)
destination(desLength)
#config(desLength)
data
```
